```python
import numpy as np 
np.set_printoptions(threshold=np.inf, linewidth=np.nan)
import pandas as pd 
import warnings 
import matplotlib.pyplot as plt 
warnings.filterwarnings('ignore') 
```

```python
!wget -P /content/sample_data https://raw.githubusercontent.com/massinoLight/tp_AlgoBD/main/villes.csv
```

```python
data_villes = pd.read_csv('/content/sample_data/villes.csv',sep=';')
data_villes
```

```python
data_villes.shape
```

```python
X=data_villes.iloc[:, 1:13].values
labels=data_villes.iloc[:, 0].values
labels

```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaler = scaler.fit_transform(X)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(x_scaler)
print(X_pca)
```

```python
import matplotlib 
plt.scatter(X_pca[:, 0], X_pca[:, 1]) 
for label, x, y in zip(labels, X_pca[:, 0], X_pca[:, 1]): 
    plt.annotate(label, xy=(x, y), xytext=(-0.2, 0.2), textcoords='offset points') 
plt.show() 
```
