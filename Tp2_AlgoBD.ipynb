```python
X=data_villes.iloc[:, 1:13].values
labels=data_villes.iloc[:, 0].values
labels

```

```python
data_crimes = pd.read_csv('/content/sample_data/crimes.csv',sep=';')
data_crimes
```

```python
data_crimes.shape
```

```python
X_crimes=data_crimes.iloc[:, 1:7].values
labels=data_crimes.iloc[:, 0].values
labels
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaler = scaler.fit_transform(X_crimes)
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(x_scaler)
print(X_pca)
```

```python
import matplotlib 
plt.scatter(X_pca[:, 0], X_pca[:, 1]) 
for label, x, y in zip(labels, X_pca[:, 0], X_pca[:, 1]): 
    plt.annotate(label, xy=(x, y), xytext=(-0.2, 0.2), textcoords='offset points') 
plt.rcParams["figure.figsize"] = 5,40    
plt.show() 
```

```python
from sklearn.cluster import KMeans
from matplotlib import pyplot

import numpy

kmeans = KMeans(n_clusters=3).fit(X_pca)
#print(len(kmeans.labels_))


pyplot.figure(figsize = (10, 10))
pyplot.title("Clustering avec kmeans ")

pyplot.xlabel("laxe des X")
pyplot.ylabel("laxe des Y")
x1 = []; y1 = []
x2 = []; y2 = []
x3 = []; y3 = []

for i in range(0,len(X_pca)):
    if kmeans.labels_[i] == 0:
        x1.append(X_pca[i][0])
        y1.append(X_pca[i][1])
    else:
        if kmeans.labels_[i] == 1:
            x2.append(X_pca[i][0])
            y2.append(X_pca[i][1])
        else:
            x3.append(X_pca[i][0])
            y3.append(X_pca[i][1])
            






pyplot.scatter(x1, y1, c = 'red')
pyplot.scatter(x2, y2, c = 'yellow')
pyplot.scatter(x3, y3, c = 'blue')


pyplot.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

clustering = AgglomerativeClustering().fit(X_pca)
clustering
clustering.labels_
```

```python
x1 = []; y1 = []
x2 = []; y2 = []
for i in range(0,len(X_pca)):
    if clustering.labels_[i] == 0:
        x1.append(X_pca[i][0])
        y1.append(X_pca[i][1])
    else:
        if clustering.labels_[i] == 1:
            x2.append(X_pca[i][0])
            y2.append(X_pca[i][1])
        
pyplot.scatter(x1, y1, c = 'red')
pyplot.scatter(x2, y2, c = 'yellow') 
pyplot.show()       
```

```python
from sklearn import metrics 
for i in np.arange(2, 6): 
    clustering = KMeans(n_clusters=i).fit_predict(X_pca) 
    print(metrics.silhouette_score(X_pca, clustering,metric='euclidean')) 
    print()
```
